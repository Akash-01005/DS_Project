```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Cleaned_Car_Details.csv")
print(df.head(10))
```

```python
first_ten = df.head(10)
last_ten = df.tail(10)

name = list()

print("\nFirst 10 Data")
for i in first_ten['name']:
  name.append(i.split(" ")[0]+'-'+i.split(" ")[1]+'-'+i.split(" ")[2])

plt.bar(name,first_ten['mileage'])
plt.xlabel("Name")
plt.ylabel("Mileage")
plt.title("Name vs Mileage")
plt.xticks(rotation=45, ha='right')
plt.show()
name.clear()

print("\nLast 10 Data")
for i in last_ten['name']:
  name.append(i.split(" ")[0]+'-'+i.split(" ")[1]+'-'+i.split(" ")[2])

plt.bar(name,last_ten['mileage'])
plt.xlabel("Name")
plt.ylabel("Mileage")
plt.title("Name vs Mileage")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import numpy as np
import re
file_path = "Car details v3.csv"
df = pd.read_csv(file_path)
df = df.drop(columns=['selling_price','seller_type','owner','km_driven'])
print(df)
df.replace(r'^\s*$', np.nan, regex=True, inplace=True)
def convert_mileage(value):
  if isinstance(value, str):
    if 'kmpl' in value:
     return float(value.replace(' kmpl', ''))
    elif 'km/kg' in value:
     return float(value.replace(' km/kg', '')) * 1.4
  return np.nan
df['mileage'] = df['mileage'].apply(convert_mileage)
df['engine'] = df['engine'].str.replace(' CC', '',
regex=False).astype(float)
def convert_max_power(value):
    if isinstance(value, str):
      match = re.search(r"(\d+\.?\d*)", value)
      return float(match.group(1)) if match else np.nan
    return np.nan
df['max_power'] = df['max_power'].apply(convert_max_power)
df[['mileage', 'engine', 'max_power']] = df[['mileage', 'engine',
'max_power']].apply(lambda x: x.fillna(x.median()))
df['seats'] = df['seats'].fillna(df['seats'].mode()[0]).astype(int)
df.loc[:, 'torque'] = df['torque'].fillna('Unknown')

df.drop_duplicates(inplace=True)
df = df.drop_duplicates(subset=['name', 'year'], keep='first')
cleaned_file_path = "Cleaned_Car_Details.csv"
df.to_csv(cleaned_file_path, index=False)
print("Data cleaning completed. Cleaned file saved as Cleaned_Car_Details.csv")
print("\n\n")
print(df)
```
