```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

df = pd.read_csv("Cleaned_Car_Details.csv")
df.head(2)
```

```python
x = df['mileage']
y = df['max_power']
plt.scatter(x, y)
plt.xlabel('Mileage')
plt.ylabel('Max Power')
plt.show()
```

```python
slope,intercept,r_value,p_value,std_err=stats.linregress(x,y)
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-value: {r_value}")
print(f"P-value: {p_value}")
print(f"Standard Error: {std_err}")

plt.scatter(x, y)
plt.plot(x,slope*x+intercept,color = 'black')
plt.xlabel('Mileage')
plt.ylabel('Max Power')
plt.title('Linear Regression')
plt.show()
```

```python
coefficients = np.polyfit(x, y, 2)
ploy_model = np.poly1d(coefficients)

x_line = np.linspace(x.min(), x.max(), 100)
y_line = ploy_model(x_line)

plt.scatter(x, y)
plt.plot(x_line, y_line, color = 'black')
plt.xlabel('Mileage')
plt.ylabel('Max Power')
plt.title('Polynomial Regression')
plt.show()

print(f"Coefficients: {coefficients}")
```

```python
r2_linear = stats.linregress(x, y).rvalue ** 2
print(f"R-squared (Linear Regression): {r2_linear}")
ploy = np.poly1d(np.polyfit(x, y, 2))
r2_poly = 1 - (np.sum((y - ploy(x)) ** 2) / np.sum((y - np.mean(y)) ** 2))
print(f"R-squared (Polynomial Regression): {r2_poly}")
if r2_linear > r2_poly:
    print("Linear Regression is better")
else:
    print("Polynomial Regression is better")
```
